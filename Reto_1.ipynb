Porcentaje Crecimiento Patrimonio: Variación porcentual del patrimonio de la persona en un período determinado, utilizada para detectar incrementos patrimoniales no explicados.


 Lavado de Activos: Es el proceso ilegal para ocultar el movimiento de dinero o bienes de una organizacion o persona.




0. Cargamos las librerias del trabajo.

```python
import numpy as np
import pandas as pd

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Cargamos la Base de Datos.

```python
nxl = "/content/2. LavadoActivos.xlsx"
XBD = pd.read_excel(nxl)
XBD.head()

XD=XBD[["edad","ingresos_usd", "gastos_usd", "num_tarjetas_credito","monto_transado_tarjetas_usd","porcentaje_crecimiento_patrimonio"]] #Variables de trabajo
XD.head(10)
yd=XBD[["lavado_activos"]]
yd.head()
```

2. Implementamos modelo Naive buyes.

```python
np.set_printoptions(suppress=True,precision=3)
mnb=GaussianNB()
mnb.fit(XD,yd)

#Mostramos las medias
u=mnb.theta_
sigma=mnb.var_;sigma=np.sqrt(sigma)
print("edad","ingresos_usd", "gastos_usd", "num_tarjetas_credito","monto_transado_tarjetas_usd","porcentaje_crecimiento_patrimonio", "lavado_activos")
print ("las medias son:")
print(u)
print ("las desviaciones son:")
print(sigma)
print('Los valores maximos son:\n',u+sigma)
print('Los valores minimos son:\n',u-sigma)
```

**Analisis de Resultado**

De los resultados arrojados por el modelo, se puede observar que la categoría de individuos que SÍ realizan lavado de activos (clase 1) presenta niveles significativamente más altos de ingresos, gastos y monto transado con tarjetas en comparación con la categoría de individuos que NO realizan lavado de activos (clase 0), lo cual evidencia la predominancia de las variables monetarias en la clasificación del modelo, ya que el lavado de activos se asocia principalmente a mayor volumen de movimiento financiero; es importante resaltar que aunque los ingresos y gastos promedio son considerablemente más altos en la clase 1, esta categoría presenta una alta dispersión en los valores, lo que indica la existencia de perfiles financieros muy diferentes dentro del grupo que sí realiza lavado, mientras que la clase 0 muestra un comportamiento más estable y homogéneo; por otro lado, la edad y el número de tarjetas de crédito presentan valores promedio muy similares entre ambas categorías, lo que sugiere que estas variables tienen un impacto limitado en la diferenciación entre individuos que realizan y no realizan lavado de activos, funcionando únicamente como variables de contexto; finalmente, el porcentaje de crecimiento patrimonial no evidencia diferencias relevantes entre las dos clases, lo que indica que, para este conjunto de datos, el modelo se apoya principalmente en la magnitud de los ingresos, gastos y transacciones como criterios determinantes para identificar posibles casos de lavado de activos, alineándose con los principios del análisis SARLAFT orientados a la detección de comportamientos financieros atípicos.
